```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("ai_job_dataset(cleaned).xlsx")
```

```python
df=df.dropna()
```

```python
plt.figure(figsize=(10,5))
ax = sns.scatterplot(
    data=df,
    x='years_experience',
    y='salary_usd',
    hue='experience_level',
    palette='mako',
    s=100,
    alpha=0.9,
    edgecolor='white'
)

plt.title("Relationship Between Salary and Years of Experience", fontsize=16, fontweight='bold', pad=15)

ax.set_xlabel('')
ax.set_ylabel('')
sns.despine(left=True, bottom=True)

plt.legend(title='', loc='upper left', bbox_to_anchor=(1,1), fontsize=10)
plt.tight_layout()
plt.show()

```

```python
top_jobs = df.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(
    x=top_jobs.values,
    y=top_jobs.index,
    palette="mako"
)

plt.title("Top 10 AI Jobs with Highest Average Salaries", fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([])

sns.despine(left=True, bottom=True)
ax.bar_label(ax.containers[0], fmt='$%.0f', fontsize=11, padding=3)

plt.tight_layout()
plt.show()
```

```python
from collections import Counter
```

```python
skills = df['required_skills'].dropna().str.split(',')  # assuming comma-separated skills
skills_list = [skill.strip() for sublist in skills for skill in sublist]
skill_counts = Counter(skills_list).most_common(10)
skills_df = pd.DataFrame(skill_counts, columns=['Skill', 'Count'])
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(
    x='Count',
    y='Skill',
    data=skills_df,
    palette='mako'
)

plt.title('Top 10 Most Required Skills in AI Jobs', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([])

sns.despine(left=True, bottom=True)
ax.bar_label(ax.containers[0], fmt='%d', fontsize=11, padding=3)

plt.tight_layout()
plt.show()
```

```python
edu_salary = df.groupby('education_required')['salary_usd'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(
    data=edu_salary,
    x='education_required',
    y='salary_usd',
    palette='mako'
)

plt.title('Average Salary by Education Level', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_yticks([])

sns.despine(left=True, bottom=True)

for container in ax.containers:
    ax.bar_label(container, labels=[f"${v:,.0f}" for v in container.datavalues],
                 fontsize=11, padding=3)

plt.xticks(rotation=30, fontsize=11)
plt.tight_layout()
plt.show()
```

```python
top_countries = df.groupby('company_location')['salary_usd'].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    palette='mako'
)

plt.title('Top 10 Countries with Highest AI Job Salaries', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([])

sns.despine(left=True, bottom=True)

ax.bar_label(ax.containers[0], labels=[f"${v:,.0f}" for v in top_countries.values],
             fontsize=11, padding=3)

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['work_type'] = df['work_type'].replace('Hybird', 'Hybrid')

avg_salary = df.groupby('work_type')['salary_usd'].mean()
order = ['On-site', 'Hybrid', 'Remote']
avg_salary = avg_salary.reindex(order).dropna()


colors = sns.color_palette("mako", len(avg_salary))

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    avg_salary.values,
    labels=avg_salary.index,
    colors=colors,
    autopct=lambda pct: f'${int(pct/100.*avg_salary.sum()):,.0f}\n({pct:.1f}%)',
    startangle=90,
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    explode=(0.05,) * len(avg_salary)
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    
plt.title('Average Salary Distribution by Work Type', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
mean_salary = (
    df.groupby('employment_type')['salary_usd']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(8, 6))
ax = sns.barplot(
    data=mean_salary,
    x='employment_type',
    y='salary_usd',
    palette='mako'
)

plt.title('Average Salary by Employment Type', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([])

sns.despine(left=True, bottom=True)

ax.bar_label(ax.containers[0], fmt='$%.0f', fontsize=11, padding=3)

plt.tight_layout()
plt.show()

```

```python
company_salary = (
    df.groupby('company_name')['salary_usd']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
    .head(5)
)

plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=company_salary,
    x='salary_usd',
    y='company_name',
    palette='mako'
)

plt.title('Top 5 Companies Offering Highest Salaries', fontsize=16, fontweight='bold')

ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticks([])

sns.despine(left=True, bottom=True)

ax.bar_label(ax.containers[0], fmt='$%.0f', fontsize=11, padding=3)

plt.tight_layout()
plt.show()
```

```python
country_hiring = df['company_location'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(
    x=country_hiring.index,
    y=country_hiring.values,
    palette='mako'
)

plt.title('Top 10 Hiring Countries', fontsize=16, fontweight='bold')
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, fontsize=11)
ax.set_yticks([])
sns.despine(left=True, bottom=True)

ax.bar_label(ax.containers[0], fontsize=11, padding=3)

plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['salary_usd', 'years_experience', 'benefits_score', 'job_description_length']

corr = df[numeric_cols].corr()

plt.figure(figsize=(8,6))
ax = sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='mako',
    linewidths=0.5,
    square=True,
    cbar_kws={'shrink': 0.8}
)

plt.title('Correlation Between Key Numeric Features', fontsize=16, fontweight='bold', pad=15)
plt.xticks(rotation=30, fontsize=11)
plt.yticks(rotation=0, fontsize=11)
plt.tight_layout()
plt.show()
```
